```python
class Library:
    
    # Creating a constructor
    def __init__(self, list, name):
        self.name = name
        self.bookslist = list
        self.library_lend = {}
        
    # Function to display the valid books in the library.
    def Displaybook(self):
        print("Hi,the available books in our library are :")
        print(self.bookslist)
              
     # Function to lend a book.
    def Lend_book(self, username, book):
        self.username = username
        self.book = book
        if book not in self.bookslist:
              print("Sorry, this book is not valid in our library")
        else:
            if book in self.library_lend.keys():
                  print("Sorry, this book is not valid now")
            else:
                self.library_lend.update({self.book : self.username})
                print("Congrats, you borrowed the book")
                      
                print("Our library dictionary is :")
                print(self.library_lend)
        
    # Function to return  a book 
    def Return_book(self, book):
        if book not in self.bookslist:
              print("Sorry, this book is not valid in our library")
        else:
            if book not in self.library_lend.keys():
                print("Sorry, you did not borrowed this book.")
            else:
                print("Thank you, you have been returned the book now.")
                del self.library_lend[book]
                
                print("Our library dictionary is :")
                print(self.library_lend)   
        
    # Function to add a book 
    def Add_book(self, newbook):
        self.newbook = newbook
        if newbook in self.bookslist:
            print("Sorry, this book is in our library")
        else:
            print("Thank you, we add the book to our library")
            self.bookslist.append(self.newbook)
     

# Real Application
Member = Library(['The Unbearable Lightness of Being','A Clockwork Orange','Tender Is the Night', 'The Haunted Vagina', 'Invisible Monsters', 'The Wasp Factory'], "READIND FOR ALL")

while(True):
    print(f"Welcome to the {Member.name} library, please enter your choice to continue")
    print("1. Display Books")
    print("2. Lend a Book")
    print("3. Return a Book")
    print("4. Add a Book")

    user_choice = input()
    if user_choice not in ['1', '2', '3', '4']:
        print("Please enter a valid choice")
        continue

    else:
        user_choice = int(user_choice)

    if user_choice == 1:
        Member.Displaybook()

    elif user_choice == 2:
        username = input("Enter your name ")
        book = input("Enter the name of the book you want to lend :")
        Member.Lend_book(username, book)

    elif user_choice == 3:
        book = input("Please, enter the book name ")
        Member.Return_book(book)

    elif user_choice == 4:
        newbook = input("Please, enter the new book name ")
        Member.Add_book(newbook)            

    else:
        print("Not a valid choice")
        print("We are happy to herp you")        
```
